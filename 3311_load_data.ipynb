```python
tv_df = pd.read_csv('./tv.csv')
movie_df = pd.read_csv('./movie.csv')

# note dateOfRelease in both TV and Movie is now release_year
tv_df = tv_df[['videoId', 'EpisodeNo','SeasonNo', 'Runtime','release_year', 'title', 'description', 'classfication']] # no need code since not specified in A3 sol
#tv_df['code'] = tv_df['videoId'].str[1:5] # since in the form sXXXXYYY where X = code, Y = total episode no
tv_df['release_year'] = tv_df['release_year'].astype(str) + '-01-01'
tv_df['Runtime_a'] = tv_df['Runtime'].str.index(' min') # get index at where substring ' min' starts.
# remove everything at or after the index computed from the previous row. lambda s with axis=1 takes s[0]=Runtime, s[1]=Runtime_a and tells us to apply slice function row-wise 
tv_df['Runtime'] = tv_df[['Runtime', 'Runtime_a']].apply(lambda s: s[0][:s[1]], axis=1).astype(int)

movie_df = movie_df[['videoId', 'sequel', 'Runtime','release_year', 'title', 'description', 'classfication']] 
movie_df['release_year'] = movie_df['release_year'].astype(str) + '-01-01'
movie_df['Runtime_a'] = movie_df['Runtime'].str.index(' min')
movie_df['Runtime'] = movie_df[['Runtime', 'Runtime_a']].apply(lambda s: s[0][:s[1]], axis=1).astype(int)

video_df = pd.concat([tv_df[['videoId', 'title', 'description', 'classfication']],
                      movie_df[['videoId', 'title', 'description', 'classfication']]])

tv_df.rename(columns={"release_year": "dateOfRelease", 'Runtime': 'runtime', 'EpisodeNo': 'episodeNo', 
                      'SeasonNo': 'seasonNo'}, inplace=True)
movie_df.rename(columns={"release_year": "dateOfRelease", 'Runtime': 'runtime'}, inplace=True)

tv_df = tv_df[['videoId', 'episodeNo','seasonNo', 'runtime','dateOfRelease']]
movie_df = movie_df[['videoId', 'sequel', 'runtime','dateOfRelease']] 

tv_df
```

```python
acts_df = pd.read_csv('./acts.csv')
acts_df = acts_df[['show_id', 'cast_id', 'role']]
acts_df.rename(columns={"show_id": "videoId", "cast_id": 'castId'}, inplace=True)
acts_df
```

```python
cast_df = pd.read_csv('./acts.csv')
cast_df = cast_df[['cast_id', 'cast_name']].drop_duplicates()
cast_df.rename(columns={"cast_id": "castId", "cast_name": 'castName'}, inplace=True)
cast_df
```

```python
tv_df.to_csv("./data/tv.csv", index=False)
movie_df.to_csv("./data/movie.csv", index=False)
video_df.to_csv("./data/video.csv", index=False)
cast_df.to_csv("./data/cast_member.csv", index=False)
acts_df.to_csv("./data/acts.csv", index=False)
```
